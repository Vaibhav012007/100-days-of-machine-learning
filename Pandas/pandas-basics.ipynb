## Series [1d]

```python
#series-1d array a sinfle column in the spreadsheet
data=[1,2,3]
series=pd.Series(data)
print(series,"\n")
series.loc[1]+=2
print(series.loc[1],"\n")

data=[1,2,3.1]
series=pd.Series(data)
print(series,"\n")

data=["A","B","C"]
series=pd.Series(data)
print(series,"\n")

data=[True, False, True]
series=pd.Series(data)
print(series,"\n")

#for custom index
data=[1,2,3]
series=pd.Series(data,index=["a","b","c"])
print(series,"\n")
print(series.loc["a"],"\n")
print(series.iloc[1],"\n")#here iloc means integer location

#here loc means locate to locate the value at the given index to access and change the value
#syntax: series_name.loc[index]
```

## Filtering for series and dataframe
keeping only the rows that match the condition

```python
##For SERIES

data=[1,2,3,4,5]
series=pd.Series(data)
print(series[series>3],"\n")

##dictionary to series example
calories={"day 1":1750,"day 2":2100,"day 3":1700}
series=pd.Series(calories)
print(series[series>1700],"\n")

##----------------------------------------------------------------------------------------------

##For DataFrame
df=pd.read_csv("C:/Projects/Machine Learning/Code/Data-Set/Pokemon_data.csv")
tall_pokemon=df[df["Height"]>=2]
heavy_pokemon=df[df["Weight"]>100]
legendary_pokemon=df[df["Legendary"]==1]
legendary_pokemon=df[df["Legendary"]==True]##same as int value of bool True is 1

water_pokemon=df[(df["Type1"]=="Water") | (df["Type2"]=="Water")] #or logical Operator
ff_pokemon=df[(df["Type1"]=="Fire") & (df["Type2"]=="Flying")]#and logical Operator
print(ff_pokemon)
```

## Dataframe [2d]

```python
data={"Name":["Mukesh","Chotu","Bablu"],
      "Age":[11,12,13],
      "Class":[4,5,6]
}
df=pd.DataFrame(data,index=["Employee 1","Employee 2","Employee 3"])
print(df,"\n")
print(df.loc["Employee 1"],"\n")
print(df.iloc[1],"\n")

#Add a new column/ series
df["Job"]=["cook","playboy","teacher"]
print(df,"\n")

#Add a new row
new_row=pd.DataFrame([{"Name":"Dimdim", "Age":26, "Class":10, "Job": "N/A"}],
                    index=["Employee 4"])
df=pd.concat([df,new_row])
print(df,"\n")


#Add a new rows
new_row=pd.DataFrame([{"Name":"Dimdim", "Age":26, "Class":10, "Job": "N/A"},
                      {"Name":"Domdim", "Age":23, "Class":12, "Job": "Auntie"},
                      {"Name":"Dimdom", "Age":25, "Class":7, "Job": "PT"}],
                    index=["Employee 5","Employee 6","Employee 7"])
df=pd.concat([df,new_row])
print(df,"\n")



# Create a sample DataFrame
data = {'Name': ['Alice', 'Bob', 'Charlie'], 'Age': [25, 30, 35]}
df = pd.DataFrame(data)

# Add a new column with a single value
df['City'] = 'New York'

# Add a new column with a list of values
df['Occupation'] = ['Engineer', 'Data Scientist', 'Manager']


#---------------------------------------------------------------------------------------------------------------
#To Remove the existing row

print(df)
# Method A: Create a new DataFrame (recommended)
df_new = df.drop(columns=['City'])
print(df_new)

# Method B: Modify the original DataFrame in-place
df.drop(columns=['City'], inplace=True)
print("\nOriginal DataFrame after dropping column 'C' in-place:")
print(df)

#Method B: Using Del Keyword which direcly deletes the column from the original array(alternative) 
## Note that this is a permanent opertion and can not be reverted so avoid using.
del df_new["Occupation"]
print(df_new)

```

## Accessing Data  with Pandas

### Selection

```python
df=pd.read_csv("C:\Projects\Machine Learning\Code\Data-Set\Iris.csv",index_col="Species")

#Selection by Column

'''print(df["Species"])
print(df["PetalLengthCm"])
print(df["PetalWidthCm"])           '''
print(df[["SepalLengthCm","SepalWidthCm"]])


## Syntax:  df.loc["row_label","column_label"]
## Syntax:  df.iloc["row_index","column_index"] 
#Selection by Row

print(df.iloc[0])#1st row
print(df.iloc[[0,1,2,3,4,5,6,7]])
print(df.loc["Iris-setosa"])#all values present in this column having this species
print(df.loc[["Iris-setosa","Iris-virginica"]])#all rows having this species
print(df.loc["Iris-setosa",["PetalWidthCm","PetalLengthCm"]])#same as above but only the selected tags will be printed

#values can be modified using loc
df.loc["Iris-setosa","SepalLengthCm"]=10
print(df)

#values can be modified using iloc
df.iloc[1,1]=10
df.iloc[1:5,1]=20 
print(df.head())
```

### Slicing

```python
#here we can apply slicing
#syntax for slicing:
#df.iloc[Rows, Columns] where for both row and columns: (start:end:step)

#print selected columns
print(df.iloc[:,0:3])

#Print selected rows
print(df.iloc[0:5:2])#prints the first 5 values and all columns are printed
print(df.loc["Iris-setosa":"Iris-versicolor":20])#all rows having this species with a gap of 20 

#print selected rows and columns
print(df.loc["Iris-setosa":"Iris-versicolor","PetalLengthCm"])
print(df.loc["Iris-setosa":"Iris-versicolor",["PetalWidthCm","PetalLengthCm"]])#same as above but only the selected tags will be printed

print(df.iloc[0:5:2,0:3])#prints the first 5 values with a step of 2 and the first 3 columns are printed
print(df.iloc[0:5:2,[0,1,2]])#prints the first 5 values with a step of 2 and the first 3 columns are printed
```

### Sorting

```python
## Sorting
print(df.sort_values("PetalWidthCm").head())
print(df.sort_values("PetalWidthCm",ascending=False).head())

print(df.sort_values(["PetalWidthCm","PetalLengthCm"]).head())
#to sort multiple lables with each lable having same order
print(df.sort_values(["PetalWidthCm","PetalLengthCm"],ascending=False).head())
#to sort multiple lables with each lable having different order
print(df.sort_values(["PetalWidthCm","PetalLengthCm"],ascending=[0,1]).head())# 0->False   1->True
```

# Important
### Functions in Pandas 

```python
df=pd.DataFrame([[1,2,3],[4,5,6],[7,8,9]],index=["A","B","C"])##to index rows
df=pd.DataFrame([[1,2,3],[4,5,6],[7,8,9]],columns=["A","B","C"])##to index columns
df=pd.DataFrame([[1,2,3],[4,5,6],[7,8,9]],columns=["A","B","C"],index=["x","y","z"])##to index both

#-------------------------------------------------------------------------------------------------------

print(df.head(10))#to see the first 5 rows
print(df.tail(10))#to see the last 5 rows
print(df.sample())#to see any random rows
print(df.sample(2,random_state=1))#this will not change (deterministic)

print(df.columns)#to see column index
print(df.index)#to see row index
print(df.shape)#to see the shape of the dataframe like numpy
print(df.size)#product of all dimensions in dataframe = total numnber of elements in df
print(df.info())#to see the info about the dataframe
print(df.describe())#to see the numerical info about the dataframe

print(df.values)#this will convert the dataframe into a numpy array
```

## Aggreagrate Functions
In pandas, aggregate functions are used to perform summary computations on data, consolidating multiple values into a single result

```python
df=pd.read_csv("C:/Projects/Machine Learning/Code/Data-Set/Pokemon_data.csv")

#Functions for whole DataFrame
print(df.describe())
print(df.mean(numeric_only=True),"\n")
print(df.sum(numeric_only=True),"\n")
print(df.min(numeric_only=True),"\n")
print(df.max(numeric_only=True),"\n")
print(df.count(),"\n")

#Funcitons for Each Columns
print(df["Height"].mean()) #to obtain mean of a particular column in a dataframe or (series)
print(df["Height"].sum())
print(df["Height"].min())
print(df["Height"].max())
print(df["Height"].count())
```
